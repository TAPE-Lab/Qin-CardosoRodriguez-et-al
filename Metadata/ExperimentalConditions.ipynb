# Mass Cytometry

## WENR Permutation

Metadata columns:
- batch 
- replicate 
- condition 
- culture
- format 
- cell-type 
- media 
- W 
- E 
- N 
- R 
- genotype
- A 
- K 
- P

```python
df = pd.read_csv("../ZenodoRepo/WENR-Permutation_all-cells.csv.zip")
metadata_columns = [
    "batch",
    "replicate",
    "condition",
    "culture",
    "format",
    "cell-type",
    "media",
    "W",
    "E",
    "N",
    "R",
    "genotype",
    "A",
    "K",
    "P",
]
df_metadata = df.loc[:, metadata_columns]
```

```python
df_metadata = df_metadata.drop_duplicates(subset=['condition']).sort_values('condition').reset_index(drop=True)
df_metadata.to_csv("SampleDF_MC_WENR.csv")
df_metadata
```
